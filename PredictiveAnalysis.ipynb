```python
#Segmentation using K-Means Clustering
```

```python
df.head()
```

```python
X=df[["age","balance","payment","pos","salary"]]
X.shape
```

```python
wcs=[]
for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit(X)
    wcs.append(km.inertia_)
```

```python
plt.figure(figsize=(12,6))
plt.plot(range(1,11),wcs)
plt.grid()
plt.plot(range(1,11),wcs, linewidth=2, color="red", marker ="8")
plt.xlabel("K Value")
plt.xticks(np.arange(1,11,1))
plt.ylabel("WCS")
plt.show()
```

```python
km=KMeans(n_clusters=3)
clusters = km.fit_predict(X)
df["label"] = clusters
```
